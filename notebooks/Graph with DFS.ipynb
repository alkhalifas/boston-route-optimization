### Limiting DFS by Number of Nodes

```python
d, q = defaultdict(list), deque([('0', 0, [])])
# print(d, "- ", q)

t, s = '13', 12 #target node and number of steps to target

d1 = {t:(0, [])}
# print(d1)

for a, *b in Graph: # change to list here
#     print(a, *b)
    d[a].append(b)
    
print(d)

while q:
    a, b, c = q.popleft()
    if a == t and len(c) == s:
        if d1[a][0] < b:
            d1[a] = (b, c)
    if len(c+[a]) <= s:
        for x, y in d[a]:
            q.append((x, b+y, c+[a]))

print(d1['13'])
```
